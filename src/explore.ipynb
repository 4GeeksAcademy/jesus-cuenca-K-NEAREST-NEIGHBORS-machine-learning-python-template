```python
import sqlite3

# Crear una conexión a la base de datos SQLite
conn = sqlite3.connect('movies_database.db')

# Almacenar los DataFrames en tablas diferentes
total_data_movies.to_sql('movies', conn, if_exists='replace', index=False)
total_data_credits.to_sql('credits', conn, if_exists='replace', index=False)

# Ver las columnas title de ambas tablas
unique_titles_movies = pd.read_sql_query("SELECT DISTINCT title FROM movies", conn)
unique_titles_credits = pd.read_sql_query("SELECT DISTINCT title FROM credits", conn)

print("Títulos únicos en movies:")
print(unique_titles_movies.head())

print("\nTítulos únicos en credits:")
print(unique_titles_credits.head())


# Ejecutar una consulta SQL para unir las dos tablas
query = '''
SELECT
    movies.*,
    credits.movie_id,
    credits.cast,
    credits.crew
FROM
    movies
INNER JOIN
    credits
ON
    credits.title = movies.title
'''

# Crear un nuevo DataFrame a partir de la consulta
unified_df = pd.read_sql_query(query, conn)

# Mostrar las primeras filas del DataFrame unificado
print("\nDataFrame Unificado:")
print(unified_df.head())

# Cerrar la conexión a la base de datos
conn.close()
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from scipy.sparse import hstack

# Asegúrate de que selected_df no tenga valores NaN en las columnas relevantes
selected_df = selected_df.fillna('')  # Llenar NaNs con cadenas vacías

# Asegurarte de que las columnas sean cadenas
selected_df['tags'] = selected_df['tags'].astype(str)
selected_df['overview'] = selected_df['overview'].astype(str)
selected_df['genres'] = selected_df['genres'].astype(str)
selected_df['keywords'] = selected_df['keywords'].astype(str)
selected_df['cast'] = selected_df['cast'].astype(str)
selected_df['crew'] = selected_df['crew'].astype(str)

# Inicializar los vectorizadores
vectorizer_tags = CountVectorizer()
vectorizer_overview = CountVectorizer()
vectorizer_genres = CountVectorizer()
vectorizer_keywords = CountVectorizer()
vectorizer_cast = CountVectorizer()
vectorizer_crew = CountVectorizer()

# Vectorizar cada columna de texto y mantener como matrices dispersas
X_tags = vectorizer_tags.fit_transform(selected_df['tags'])
X_overview = vectorizer_overview.fit_transform(selected_df['overview'])
X_genres = vectorizer_genres.fit_transform(selected_df['genres'])
X_keywords = vectorizer_keywords.fit_transform(selected_df['keywords'])
X_cast = vectorizer_cast.fit_transform(selected_df['cast'])
X_crew = vectorizer_crew.fit_transform(selected_df['crew'])

# Combinar todas las matrices de características en una sola
X = hstack((X_tags, X_overview, X_genres, X_keywords, X_cast, X_crew))


```

```python
from sklearn.neighbors import NearestNeighbors

# Inicializar el modelo NearestNeighbors
knn = NearestNeighbors(n_neighbors=5, metric='cosine')  # Puedes ajustar n_neighbors según sea necesario

# Ajustar el modelo a la matriz de características
knn.fit(X)

# Ejemplo: Encontrar películas similares para la primera película
# Puedes cambiar el índice para buscar películas similares a otras
distances, indices = knn.kneighbors(X[15], n_neighbors=5)  # Busca las 5 más cercanas a la primera película

# Mostrar las películas similares
print("Índices de las películas similares:", indices)
print("Distancias a las películas similares:", distances)

# Para mostrar títulos, asegúrate de tener un DataFrame que contenga títulos
similar_movies = selected_df.iloc[indices[0]]  # Obtiene los datos de las películas similares
print("\nPelículas Similares:")
print(similar_movies[['title']])

```

```python
def recommend(movie, knn_model, data_frame, n_recommendations=5):
    # Verificar si la película está en el DataFrame
    if movie not in data_frame['title'].values:
        print(f"Movie '{movie}' not found in the dataset.")
        return

    # Obtener el índice de la película que coincide con el título
    movie_index = data_frame[data_frame['title'] == movie].index[0]

    # Obtener la representación de características de la película
    movie_features = X[movie_index].reshape(1, -1)  # Asegúrate de usar la matriz de características

    # Obtener los índices y distancias de las películas más cercanas
    distances, indices = knn_model.kneighbors(movie_features, n_neighbors=n_recommendations + 1)

    # Mostrar las películas recomendadas
    print(f"\nTop {n_recommendations} movies similar to '{movie}':")
    for i in range(1, len(indices.flatten())):  # Comenzar desde 1 para evitar la película misma
        print(data_frame.iloc[indices.flatten()[i]]['title'])

# Ejemplo de uso:
recommend('The Chronicles of Narnia: Prince Caspian', knn, selected_df)
```
